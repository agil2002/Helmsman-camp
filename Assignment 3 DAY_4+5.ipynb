```
students.plot.scatter("writing score","reading score",figsize=(9,3))
```

This is a scatter plot with the writing score and reading scores of different students

*From the above plots we can understand that the students who are good in writing is also the same as in reading and a very very few students who are  better in the reading score is weaker in writing as compared to their reading scores. 
Even though if we look in total who is good in reading is good in writing and who are weak in reading is weak in writing.*

```
import seaborn as sns
import matplotlib.pyplot as plt

xData = [x for x in range(10)]
yData = [y for y in range(20)]
sns.lineplot(x="writing score",y="reading score",data=df)
plt.show()
```

This is a line plot of the writing and reading score of the students

*From this plot too we get to know that those who are good in writing is good in reading.Those who are not good in reading is not good in writng.
